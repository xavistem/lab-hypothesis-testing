In this challenge, we will be working with pokemon data. The data can be found here:

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import levene
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
# Dragon vs Grass en HP

""""
1) Formulación de hipótesis
H0: media_HP(Dragon) = media_HP(Grass)
H1: media_HP(Dragon) > media_HP(Grass)
"""

# 2) Extracción de muestras
dragon_hp = df[df["Type 1"] == "Dragon"]["HP"]
grass_hp  = df[df["Type 1"] == "Grass"]["HP"]

# 3) Nivel de significancia
alpha = 0.05

# 4) Comprobación de condiciones
# n_dragon, n_grass > 30 → asumimos normalidad por CLT
# Homogeneidad de varianzas con Levene
stat_le, p_le = levene(dragon_hp, grass_hp)
print("Levene p‑value:", p_le)

equal_var = (p_le > alpha)  # True si varianzas iguales

# 5) Test estadístico: Two‑Sample t‑Test (one‑tailed, 'greater')
t_stat, p_value = st.ttest_ind(dragon_hp, grass_hp, equal_var=equal_var, alternative="greater")
print("t‑stat:", t_stat, "p‑value:", p_value)

# 6) Decisión e interpretación
if p_value < alpha:
    print("Rechazamos H0: los Dragon tienen más HP que los Grass")
else:
    print("No rechazamos H0: no hay evidencia suficiente")
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
# Legendary vs No Legendary en 6 estadísticas

""""
1) Formulación de hipótesis (para cada stat)
H0: media_stat(Legendary) = media_stat(NoLegendary)
H1: media_stat(Legendary) ≠ media_stat(NoLegendary)
"""

# 2) Extracción de muestras
legend = df[df["Legendary"] == True]
non_leg = df[df["Legendary"] == False]

stats_columns = ["HP","Attack","Defense","Sp. Atk","Sp. Def","Speed"]

# 3) Nivel de significancia
alpha = 0.05

## 4) Bucle para cada stat, comprobamos varianzas y t‑test

resultados = []
for col in stats_columns:
    # Levene
    _, p_le = levene(legend[col], non_leg[col])
    eq_var = (p_le > alpha)
    # Two-sample t-test (two‑tailed)
    t_s, p_v = st.ttest_ind(legend[col], non_leg[col], equal_var=eq_var)
    decision = "Rechazar H0" if p_v < alpha else "No rechazar H0"
    resultados.append({
        "Stat": col,
        "Levene_p": round(p_le,4),
        "EqualVar": eq_var,
        "T_stat": round(t_s,3),
        "P_value": round(p_v,4),
        "Decisión": decision
    })

pd.DataFrame(resultados)
```

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df2.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
# Proximidad a escuela u hospital vs precio de la casa

""""
1) Formulación de hipótesis
H0: media_price(Close) = media_price(Far)
H1: media_price(Close) ≠ media_price(Far)
"""

# 2) Función de distancia
def euclid(x1, y1, x2, y2):
    return sqrt((x1 - x2)**2 + (y1 - y2)**2)

# 3) Funciones para apply (escuela y hospital)
def distancia_escuela(row):
    return euclid(row["longitude"], row["latitude"], -118, 34)

def distancia_hospital(row):
    return euclid(row["longitude"], row["latitude"], -122, 37)

# 4) Crear las columnas de distancia usando apply
df2["dist_school"]   = df2.apply(distancia_escuela, axis=1)
df2["dist_hospital"] = df2.apply(distancia_hospital, axis=1)

# 5) Marcar 'Close' si la distancia a escuela o hospital < 0.5
df2["Close"] = (df2["dist_school"] < 0.5) | (df2["dist_hospital"] < 0.5)

# 6) Separar los valores de precio
close_vals = df2.loc[df2["Close"], "median_house_value"]
far_vals   = df2.loc[~df2["Close"], "median_house_value"]

# 7) Nivel de significancia
alpha = 0.05

# 8) Comprobación de homogeneidad de varianzas (Levene)
stat_le, p_le = levene(close_vals, far_vals)
print("Levene p‑value:", p_le)
equal_var = (p_le > alpha)

# 9) Two‑Sample t‑Test (two‑tailed)
t_stat, p_value = st.ttest_ind(close_vals, far_vals, equal_var=equal_var)
print("t‑stat:", t_stat, "p‑value:", p_value)

# 10) Decisión e interpretación
if p_value < alpha:
    print("Rechazamos H0: las casas cercanas son significativamente diferentes en precio.")
else:
    print("No rechazamos H0: no hay evidencia de diferencia en precio por proximidad.")
```
